```python
import pandas as pd
import datetime
data1 = pd.read_csv('news_total_0626.csv')
data1
```

```python
data1 = data1.drop(columns='ym')
data1.date = pd.to_datetime(data1.date,format='%Y/%m/%d')
data1['year'] = data1.date.dt.year
data1['month'] = data1.date.dt.month
data1['day'] = data1.date.dt.day
data1.iloc[:,3:4].describe()
```

```python
data2 = pd.read_csv('news_newslen.csv')
data2.title = data2.title.str.replace(' ', '')
data2['date'] = data2.time.str.replace(' ', '')
data2 = data2.drop(columns='time')
data2.content = data2.content.str.replace(' ', '')
data2 = data2[['title','date','content']]
```

```python
# 去除國際新聞
data2 = data2[~data2['date'].str.contains('國際')]
```

```python
data2.date = data2.date.str[:10]
#data2.describe(include=['object'])
data2
```

```python
data2.date = pd.to_datetime(data2.date,format='%Y/%m/%d')
data2['source'] = 'newslen'
data2['year'] = data2.date.dt.year
data2['month'] = data2.date.dt.month
data2['day'] = data2.date.dt.day
data2.head()
```

```python
data2.describe(include=['object'])
```

```python
# combining data1 & data2
news_data = pd.concat([data1,data2])
news_data.describe(include=['object'])
news_data.source = news_data.source.astype(str)
```

```python
news_data
news_data.source.unique()
```

```python
news_data.iloc[2500:2570,:]
```

```python
news_data = news_data[['title','date','year','month','day','source','content']]
```

```python
news_data.to_csv('news_total_0703_no_inter.csv',encoding='utf-8',index=False)
```
